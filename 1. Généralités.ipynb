## 1.1. Un peu d'histoire

La genèse de cette langue remonte à la fin des années 80. [Guido van Rossum](https://fr.wikipedia.org/wiki/Guido_van_Rossum), qui travaillait à l'époque à l'Institut néerlandais de recherche en mathématiques et en informatique ([CWI](http://www.cwi.nl)) à Amsterdam. Guido van Rossum a publié la version 0.9.0 de l'interpréteur en février 1991. Il travaille maintenant pour Microsoft après avoir travaillé pour d'autres entreprises telles que Dropbox et Google.

## 1.2. Que signifie un langage de programmation interprété ?

- Ordinateur $\rightarrow$ CPU $\rightarrow$ jeu d'instructions (ISA) $\rightarrow$ langage binaire.
- Un langage de programmation interprété permet d'écrire des programmes dans des langues plus adaptées à l'homme, mais nécessite une étape de traduction.
- L'exemple peut être une langue étrangère, nous avons besoin d'un traducteur ou d'un interprète.
  - Le traducteur lira le texte et produira une version en langue étrangère.
  - L'interprète, tout en lisant le texte, effectue une traduction instantanée.
- Pour un langage informatique, c'est à peu près la même chose. Il existe des [compilateurs](https://fr.wikipedia.org/wiki/Compilateur) et des [interprètes](https://fr.wikipedia.org/wiki/Interpr%C3%A8te_(informatique))).
  - Les compilateurs traduisent tout le code source en un langage binaire qui peut être utilisé directement par la CPU (unité centrale).
  - L'interpréteur lit une partie du code source et exécute directement les instructions binaires correspondantes.

### 1.3.2. Intérêt pour les sciences

Pour l'utilisation scientifique, Python est intéressant à plusieurs égards. En effet, il est capable d'exécuter automatiquement et efficacement un certain nombre de tâches qui font le quotidien des scientifiques :
- Manipuler et traiter des données provenant de simulations ou d'expériences.
- Visualiser les résultats.
- Communiquer les résultats sous forme de données numériques formatées, de figures ou d'animations.
- Dans le domaine de **l'informatique scientifique**, Python est particulièrement riche en fonctionnalités grâce à l'importante contribution de la communauté mathématique et informatique à travers le projet [SciPy](http://www.scipy.org/).


### 1.3.3. Intérêt pour l'automatisation

Le langage Python a rendu la programmation plus accessible. Il est conçu pour fonctionner sur des ordinateurs et sur des machines disposant d'un niveau considérable de traitement, de mémoire et d'autres ressources périphériques. Pour les systèmes embarqués soumis à des contraintes d'interface et de ressources plus strictes (par exemple, les radars, les systèmes d'éclairage, le contrôle d'accès), une version optimisée pour les microcontrôleurs, appelée [MicroPython](https://micropython.org/), a gagné en popularité.

La portée de cette version est croissante et a conduit la communauté open source à porter MicroPython sur les microcontrôleurs.

### 1.3.4. CircuitPython : Une version de Python chez les cartes de développement Adafruit

[Adafruit Industries](https://www.adafruit.com/) est le fournisseur de circuits électroniques choisi pour ce cours. L'entreprise conçoit, fabrique et vend des produits électroniques et de composants électroniques, des outils et des accessoires via son commerce en ligne.

Chez Adafruit, vous pouvez retrouver beaucoup de circuits imprimés (*shields*) spécifiques pour les cartes Arduino ou Raspberry PI mas aussi une grande variété de cartes de développement (*breakout*) et des créations originales.

Ce fournisseur a investi du temps dans [MicroPython](https://micropython.org/) et a dérivé une version sous le nom [CircuitPython](https://learn.adafruit.com/welcome-to-circuitpython/what-is-circuitpython). CircuitPython facilite le prototypage électronique de sorte qu'il est aussi facile de faire du prototypage de circuit avec MicroPython/CircuitPython qu'avec Arduino.

Probablement, vous êtes familier avec les produits Arduino et son environnement de développement [IDE Arduino](https://www.arduino.cc/en/software). Avec les cartes de développement Adafruit, vous pouvez aussi utiliser ces connaissances de base en Arduino. 

Dans la partie exploratoire et laboratoire du cours, plusieurs des produits Adafruit seront utilisés. Le kit d'apprentissage sera décrit au chapitre 13 du cours. En tant qu'étudiant inscrit au cours  [GAE-1003 Énergie, contrôle et instrumentation](https://www.ulaval.ca/etudes/cours/gae-1003-energie-controle-et-instrumentation), vous aurez accès au kit. Sinon, vous devez acheter les pièces par vos propres moyens.
