## Moulage

Il peut y avoir des moments où vous voulez spécifier un type sur une variable. 
Le moulage en python se fait donc à l'aide de fonctions constructeur :

int() - construit un nombre entier à partir d'un nombre entier littéral, d'un nombre flottant littéral (en arrondissant au nombre entier précédent), ou d'un nombre entier de chaîne (à condition que la chaîne représente un nombre entier)

float() - construit un nombre flottant à partir d'un entier littéral, d'un entier littéral ou d'une chaîne de caractères (à condition que la chaîne représente un entier ou un flottant)

str() - construit une chaîne de caractères à partir d'une grande variété de types de données, y compris des chaînes de caractères, des entiers littéraux et des flottants.

````python
x = int(10)   
y = int(11.8) 
z = int("12")

print(x,y,z)
````

````python
x = float(20)
y = float(21.2)
z = float("22")

print(x,y,z)
````

````python
x = str("chaîne")
x = "chaîne"

y = str(31)
z = str(32.3)

print("Voilà : " + str(45))

print(x,y,z)
````

## Commentaires sur Python

Les commentaires sont utilisés pour expliquer le code, aussi pour rendre le code plus lisible et, finalément, pour empêcher l'exécution lors du test du code.

Le commentaire commence par #, et Python les ignore.


````python
# Un commentaire
print('Bonjour!')
````

Les commentaires peuvent être placés à la fin d'une ligne. Python ignore le contenu à partir de # 

````python
print('Bonjour!') # Un commentaire
````

Voici un exemple afin d'ignorer une partie du code.


````python
#print('Bonjour!')
print('Bonsoir!')
````

Python n'a pas vraiment de syntaxe pour les commentaires multi-lignes.

Pour ajouter un commentaire multiligne, vous pouvez insérer un # pour chaque ligne, ou, vous pouvez utiliser une chaîne multiligne.

````python
#Ceci est un commentaire
#écrit en
#plus d'une ligne
print('Bonjour')

'''
Ceci est un commentaire
écrit en
plus d'une ligne
'''
print('Bonjour!')
````

## Conditionnels et côntrole de flux

Bien entendu, on peut utiliser Python pour des tâches plus compliquées que d’additionner deux et deux.

L'instruction ```for``` de Python itère sur les éléments d’une séquence (qui peut être une liste, une chaîne de caractères…), dans l’ordre dans lequel ils apparaissent dans la séquence.

````python
x = [1, 'a', 2, 'b', 3.3]

for item in x:
    print(item)
print(2+2)
````

La boucle ```while``` s’exécute tant que la condition (ici : ```inférieur ou égal à la longueur de la chaîne x```) reste vraie. En Python, comme en C, tout entier différent de zéro est vrai et zéro est faux. La condition peut aussi être une chaîne de caractères, une liste, ou en fait toute séquence ; une séquence avec une valeur non nulle est vraie, une séquence vide est fausse.

Le corps de la boucle est indenté au même niveau : l’indentation est la méthode utilisée par Python pour regrouper des instructions.

````python
i=0
while( i < len(x) ):
    print(x[i])
    i = i + 1
````

L'instruction ```if .. else``` est sans doute la plus connue.

Il peut y avoir un nombre quelconque de parties ```elif``` et la partie else est facultative. Le mot clé ```elif``` est un raccourci pour *else if*, mais permet de gagner un niveau d’indentation.

````python
food = 'poutine'

if food == 'poutine':
    print('Ummmm, ma préférée!')
    #('J'ai envie de le dire 20 fois...')
    print("""J'ai envie de le dire 100 fois...""")
    print(20 * (food + '! '))
````

````python
food = 'pomme'

if food == 'poutine':
    print('Ummmm, ma préférée!')
else:
    print('Non. Je veux de la poutine!')
````
